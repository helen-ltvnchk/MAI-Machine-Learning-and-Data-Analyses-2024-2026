Квантили   и   процентные   точки  случайной   величины. Привести пример расчета 10% проц.точки на небольшом датасете (10-20 значений).

Квантили и процентные точки случайной величины - это значения, которые делят распределение случайной величины на определённые части.

Квантиль - значение, которое делит распределение на 2 части: одна часть лежит ниже этого значения, вторая - выше. 

Квантиль n-го порядка - это значение х, для которого:
P(X <= x(q)) = q,
где q - вероятность (0.1, 0.5, 0.8 и т.д.)

Нижний квартиль - это квантиль 0.25
Медиана - это квантиль 0.5
Верхний квартиль - это квантиль 0.75.

Основные типы квантилей:
- медиана - квантиль 0.5
- Квартили - Первый - 0.25, второй - 0,5, третий - 0,75
- Децили - Делят данные на 10 равных частей. 
- Перцентили - Делят данные на 100 равных частей.

Процентные точки - это квантили, выраженные в процентах.
Например: 10%-я точка соответствует квантилю уровня 0.1, 40% - квантилю 0.4

```python
# Пример рассчёта 10%-й процентной точки на небольшом датасете

import numpy as np
import random

# Сгенерируем датасет на 20 значений, зафиксируем его
random.seed(42)
data = [random.randint(1, 50) for _ in range(20)]
print(data)

# 10%-я точка (квантиль 0.1)
quantile_01 = np.percentile(data, 10)
print(f"10%-я точка: {quantile_01:.1f}")
```
